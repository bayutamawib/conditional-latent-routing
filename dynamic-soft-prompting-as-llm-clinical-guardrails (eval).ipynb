```python
from IPython.display import HTML, display
import base64
import os

def create_download_link(file_path, title="Download Evaluation Results (Hybrid)"):
    if not os.path.exists(file_path):
        return HTML(f"<b style='color:red;'>❌ File '{file_path}' belum ditemukan. Pastikan evaluate.py sudah selesai dijalankan.</b>")
        
    with open(file_path, 'rb') as f:
        data = f.read()
    b64 = base64.b64encode(data)
    payload = b64.decode()
    html = f'''<a download="evaluation_results_hybrid.json" href="data:text/json;base64,{payload}" target="_blank">
               <button style="padding:12px 24px; background-color:#28a745; border:none; border-radius:8px; color:white; cursor:pointer; font-size:16px; font-weight:bold;">
               ⬇️ {title}</button></a>'''
    return HTML(html)

# Eksekusi pembuatan tombol
display(create_download_link('/kaggle/working/evaluation_results.json'))
```

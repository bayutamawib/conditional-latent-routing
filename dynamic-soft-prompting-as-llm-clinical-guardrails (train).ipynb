```python
import os
import shutil

# 1. Menambal config.py (Melebarkan Soft Tokens 32 -> 128)
with open('/kaggle/working/config.py', 'r') as f:
    kode_config = f.read()
kode_config = kode_config.replace('NUM_SOFT_TOKENS = 32', 'NUM_SOFT_TOKENS = 128')
with open('/kaggle/working/config.py', 'w') as f:
    f.write(kode_config)

# 2. Menambal inference_pipeline.py (Memaksa format [INST] Mistral)
with open('/kaggle/working/inference_pipeline.py', 'r') as f:
    kode_infer = f.read()
kode_infer = kode_infer.replace('f"### System:\\n{SLOW_LANE_SYSTEM_PROMPT}\\n\\n"', 'f"[INST] {SLOW_LANE_SYSTEM_PROMPT}\\n\\n"')
kode_infer = kode_infer.replace('f"### Patient Record:\\n{raw_text}\\n\\n"', 'f"Patient Record:\\n{raw_text}\\n\\n"')
kode_infer = kode_infer.replace('f"### Assessment:\\n"', 'f"Assessment: [/INST]"')
with open('/kaggle/working/inference_pipeline.py', 'w') as f:
    f.write(kode_infer)

# 3. Menghapus otak/checkpoint lama agar tidak tabrakan dimensi saat training ulang
folder_ckpt = '/kaggle/working/checkpoints'
folder_final = '/kaggle/working/final_models'
if os.path.exists(folder_ckpt): shutil.rmtree(folder_ckpt)
if os.path.exists(folder_final): shutil.rmtree(folder_final)

print("✅ KODE BERHASIL DI-PATCH DAN MODEL LAMA TELAH DIBERSIHKAN!")
print("Siap untuk Training Ulang.")
```

```python
with open('/kaggle/working/train.py', 'r') as f:
    lines = f.readlines()

with open('/kaggle/working/train.py', 'w') as f:
    masuk_fungsi_validate = False
    for line in lines:
        if 'def validate(' in line:
            masuk_fungsi_validate = True
            
        # Jika ketemu baris ini dan posisinya sedang di dalam fungsi validate
        if 'noise_logits, z = encoder(**enc_tokens)' in line and masuk_fungsi_validate:
            # Kembalikan ke 8 spasi
            f.write('        noise_logits, z = encoder(**enc_tokens)\n')
        else:
            f.write(line)

print("✅ Spasi di fungsi validate sudah dikembalikan menjadi 8 spasi!")
```

```python
import os
import glob
from pathlib import Path

ckpt_dir = '/kaggle/working/checkpoints/'
ckpts = sorted(glob.glob(os.path.join(ckpt_dir, '*.pt')))

# 1. Hapus file terakhir yang korup (crash di tengah jalan)
if len(ckpts) > 0:
    print(f"🗑️ Menghapus file korup: {ckpts[-1]}")
    os.remove(ckpts[-1])

# 2. Sisakan SATU file yang paling sehat, buang file awal-awal
ckpts = sorted(glob.glob(os.path.join(ckpt_dir, '*.pt')))
while len(ckpts) > 1:
    print(f"🗑️ Membersihkan file lama: {ckpts[0]}")
    os.remove(ckpts[0])
    ckpts = sorted(glob.glob(os.path.join(ckpt_dir, '*.pt')))

if len(ckpts) == 1:
    print(f"✅ AMAN! File sehat yang siap di-resume: {os.path.basename(ckpts[0])}")

# 3. Patch train.py untuk Auto-Delete file lama setiap kali selesai save
with open('/kaggle/working/train.py', 'r') as f:
    kode = f.read()

# Menyuntikkan perintah hapus file lama ke dalam fungsi save_checkpoint
patch_auto_delete = """    print(f"  📁 Checkpoint saved: {path}")
    
    # Auto-delete checkpoint lama agar 20GB Kaggle tidak meledak
    ckpts_lama = sorted(Path(checkpoint_dir).glob("checkpoint_epoch_*.pt"))
    for old_file in ckpts_lama[:-1]:
        os.remove(old_file)
        
    return path"""

kode = kode.replace(
    '    print(f"  📁 Checkpoint saved: {path}")\n    return path', 
    patch_auto_delete
)

with open('/kaggle/working/train.py', 'w') as f:
    f.write(kode)

print("✅ train.py sudah di-upgrade dengan sistem Auto-Delete!")
```
